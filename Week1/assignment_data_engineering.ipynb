```python
# function to calculate great circle distance between two locations
def calc_gcd(r):
    lat1 = np.radians(r.lat_orig)
    lon1 = np.radians(r.lon_orig)
    lat2 = np.radians(r.lat_dest)
    lon2 = np.radians(r.lon_dest)

    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = np.sin(dlat / 2) ** 2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2) ** 2
    c = 2 * np.arcsin(np.sqrt(a))
    d = c * 6371 * 0.539957  # nautical miles
    return d
```

## Query flights from June 15th, and then calculate the great circle distance with apply()

```python
flights.query("aat.dt.day == 15 and aat.dt.month == 6").apply(calc_gcd, axis=1)
```

### With one nest command: 
- add a new column named **gcd** (great circle distance) for all the flights
- then add another new column that calculates the different between **distance** and **gcd**, with column name **distance_diff**

```python
flights1 = flights.assign(gcd=lambda d: d.apply(calc_gcd, axis=1)).assign(
    distance_diff=lambda d: d.distance - d.gcd
)
flights1
```

```python
# does the same, some shorter code
flights1 = flights.assign(
    gcd=lambda x: calc_gcd(x), distance_diff=lambda x: x.distance - calc_gcd(x)
)
flights1
```

### Nesting operations

- Query flights originated from Paris Charles de Gaulle Airport (airport code: LFPG)
- Calculate arrival delays of all airports
- Group by destination
- Calculated the mean flight distance and average arrival delay at the destination airports
- Sow and sort results for destinations with average arrival delay large than 10 minutes with more than 5 flights

```python
flights2 = (
    flights1.query("origin=='LFPG'")
    .eval("arrival_delay = aat - fat")
    .groupby("destination")
    .agg({"distance": "mean", "arrival_delay": "mean", "origin": "count"})
    .reset_index()
    .rename(columns={"origin": "count"})
    .query(
        "count>5 and arrival_delay.dt.components.days>-1 and arrival_delay.dt.components.minutes>10"
    )
    .sort_values(by='arrival_delay', ascending=False)
)
flights2
```

### Plot the departure and arrival delay trends over the 24-hours for at Paris Charles de Gaulle Airport (LFPG) airport
-  based on all flights departing and arriving at LFPG
- you should consider the variations for different days

```python
flightsParis = flights.query("origin == 'LFPG' or destination == 'LFPG'").assign(
    delay=lambda x: (x.fat - x.aat).dt.total_seconds() / 60
)
flightsParis = flightsParis.assign(fat_hour=flightsParis["fat"].dt.hour)
flightsParis = flightsParis.assign(fat_day=flightsParis["fat"].dt.day)

fig, ax = plt.subplots(1, figsize=(11, 6))
sns.lineplot(data=flightsParis, x="fat_hour", y="delay", hue="fat_day")
ax.set_xlabel("Hour", fontsize=18)
ax.set_ylabel("Delay (min.)", fontsize=18)
plt.legend(title="Day of the month")
plt.show()
```

### Using boxplot() to plot the arrival and departure delay statistics for each of the top 15 airport 
 - remark: top airports according to number of total departures

```python
# get 15 top airports
topAirports = (
    flights.groupby("origin")
    .size()
    .sort_values(ascending=False)
    .iloc[:15]
    .index.to_list()
)

topAirports = flights.assign(
    delay_departure=lambda x: (x.aobt - x.fobt).dt.total_seconds() / 60,
    delay_arrival=lambda x: (x.aat - x.fat).dt.total_seconds() / 60,
).query("origin in @topAirports")

# melted the delays in the same column with another column seperating the values into variable names
topAirports_ = pd.melt(
    topAirports,
    id_vars=["origin"],
    value_vars=["delay_arrival", "delay_departure"],
    var_name="delay_type",
    value_name="delay",
)
```

```python
fig, ax = plt.subplots(1, figsize=(11, 6))
sns.boxplot(data=topAirports_, x="origin", y="delay", hue="delay_type")
sns.set(font_scale=1)
ax.set_xlabel("Airport", fontsize=18)
ax.set_ylabel("Delay (min.)", fontsize=18)
plt.show()
```

### Plot the same result from previous visualization in violinplot()
- remember to try different **split**, **palette**, and **inner** arguments

```python
fig, ax = plt.subplots(1, figsize=(11, 6))
sns.violinplot(
    data=topAirports_,
    x="origin",
    y="delay",
    hue="delay_type",
    split=True,
    inner="quart",
)
sns.set(font_scale=1)
ax.set_xlabel("Airport", fontsize=18)
ax.set_ylabel("Delay (min.)", fontsize=18)
plt.show()
```
